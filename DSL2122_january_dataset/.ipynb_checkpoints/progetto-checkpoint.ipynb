```python
def replace_pattern(tweets: pd.DataFrame) -> pd.DataFrame:
    
    # regex pattern
    hashtags = "#[\d\w]+"
    mentioned = "@[\d\w]+"
    ampersand = "&[\d\w]+"
    urls = '((www.[^s]+)|(https?://[^s]+))'

    pat = hashtags + "|" + mentioned +  "|" + ampersand + "|" + urls
    
    repl = " "
    
    tweets['text'] = tweets['text'].str.replace(pat= pat, repl = repl, regex=True)

    return tweets
```
